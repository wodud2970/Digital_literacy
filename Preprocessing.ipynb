```python
import pandas as pd
data = pd.read_csv('./data/all_data(yes_star).csv', encoding='utf-8')
data
```

```python
# * 제거
lst = []

for j in range(0,data.shape[1]):
    for i in range(0,data.shape[0]):
        if data.iloc[i,j] == "*":
            if i not in lst: # 중복값 제거
                lst.append(i)    
                
for i in lst:
    data = data.drop(i,axis = 0)
data.reset_index(drop=True,inplace=True)
```

```python
data.to_csv('./data/all_data(no_star).csv')
```

```python
#1년간_교육
teaching_list = []
for i in range(len(data)):
    if data['Q8A4'][i] in [3,4] or data['Q8B4'][i] in [3,4]:
        teaching_list.append(1) #있다
    else:
        teaching_list.append(0)

data = data.drop(columns = ['Q8A4','Q8B4'])
data['1년간_교육'] = teaching_list

data['1년간_교육'].unique(), data['1년간_교육'].shape
```
