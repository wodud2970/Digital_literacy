```python
import pandas as pd

df = pd.read_csv("./data/2020_노인실태조사_부산.csv",encoding="EUC-KR")
```

```python
columns = ["gu", "B1", "C10", "D5", "D12_1_3", "D12_1_1", "D12_1_2", "D13_3", "D13_5", "D13_8","D13_9", "D13_11", "E1", "EDM", "F13", "G1_1", "G3_1", "G4_1", "H16_2", "H16_1", "H16_5", "H16_7", "J3b_3_13","Q1", "RES_SEX", "RES_EDU1"]

df = df[columns]
```

```python
df.info()
```

```python
# B1무응답 결측치 삭제
df = df[df["B1"] != 9]
df.reset_index(drop=True,inplace=True)
df["B1"].value_counts()
```

```python
# D13_3 무응답 결측치 삭제
df = df[df["D13_3"] != 9]
df.reset_index(drop=True,inplace=True)
df["D13_3"].value_counts()
```

```python
# 있다 1 없다 0으로 인스턴스 수정
df["C10"] = df["C10"].apply(lambda x: 1 if x==1 else 0)
df["D5"] = df["D5"].apply(lambda x: 1 if x==1 else 0)
df["D12_1_3"] = df["D12_1_3"].apply(lambda x: 1 if x==1 else 0)
df["D12_1_1"] = df["D12_1_1"].apply(lambda x: 1 if x==1 else 0)
df["D12_1_2"] = df["D12_1_2"].apply(lambda x: 1 if x==1 else 0)
df["D13_3"] = df["D13_3"].apply(lambda x: 1 if x==1 else 0)
df["D13_5"] = df["D13_5"].apply(lambda x: 1 if x==1 else 0)
df["D13_8"] = df["D13_8"].apply(lambda x: 1 if x==1 else 0)
df["D13_9"] = df["D13_9"].apply(lambda x: 1 if x==1 else 0)
df["D13_11"] = df["D13_11"].apply(lambda x: 1 if x==1 else 0)
df["F13"] = df["F13"].apply(lambda x: 1 if x==1 else 0)
df["G1_1"] = df["G1_1"].apply(lambda x: 1 if x==1 else 0)
df["G3_1"] = df["G3_1"].apply(lambda x: 1 if x==1 else 0)
df["G4_1"] = df["G4_1"].apply(lambda x: 1 if x==1 else 0)
df["RES_SEX"] = df["RES_SEX"].apply(lambda x: 1 if x==1 else 0)
```

```python
from sklearn.preprocessing import MinMaxScaler

scalar = MinMaxScaler(feature_range =(1, 4))
df["B1"] = scalar.fit_transform(df["B1"].to_numpy().reshape(-1,1))
df["H16_2"] = scalar.fit_transform(df["H16_2"].to_numpy().reshape(-1,1))
df["H16_1"] = scalar.fit_transform(df["H16_1"].to_numpy().reshape(-1,1))
df["H16_5"] = scalar.fit_transform(df["H16_5"].to_numpy().reshape(-1,1))
df["H16_7"] = scalar.fit_transform(df["H16_7"].to_numpy().reshape(-1,1))
```

```python
# J3b_3_13
df = df[(df["J3b_3_13"] != 999998) | (df["J3b_3_13"] != 999999)]
df.reset_index(drop=True,inplace=True)
```

```python
lst = []

for i in df["E1"]:
    if i == 1:
        lst.append(1)
    elif i in (2, 3) :
        lst.append(0)
        
df["E1"] = lst
df["E1"].value_counts()
```

```python
df = df[df["RES_EDU1"] != 9]
df.reset_index(drop=True,inplace=True)

lst = []

for i in df["RES_EDU1"]:
    if i in (0,1,2,3):
        lst.append(1)
    elif i == 4 :
        lst.append(2)
    elif i == 5 :
        lst.append(3)
    else:
        lst.append(4)
        
df["RES_EDU1"] = lst
df["RES_EDU1"].value_counts()
```

```python
from scipy.stats import mode

gu=df['gu'].unique()
Q1 = []
for i in gu:
    mode = df[df['gu'] == i]['Q1'].mode()
    if len(mode)>= 2:
        c = [x for x in mode]
        Q1.append(c)
        continue
    Q1.append(mode[0])
Q1
```

```python
gu=df['gu'].unique()
RES_EDU1 = []
for i in gu:
    mode = df[df['gu'] == i]['RES_EDU1'].mode()
    if len(mode)>= 2:
        c = [x for x in mode]
        RES_EDU1.append(c)
        continue
    RES_EDU1.append(mode[0])
RES_EDU1
```

```python
# 최빈값
columns = ["gu","C10", "D5", "D12_1_3", "D12_1_1", "D12_1_2", "D13_3", "D13_5", "D13_8", "D13_11", "E1", "EDM", "F13", "G1_1", "G3_1", "G4_1","Q1", "RES_SEX", "RES_EDU1"]
df_group = df[columns].groupby("gu")
df_group_1 = df_group.agg(lambda x:x.mode())
df_group_1
```

```python
df_group_1['RES_EDU1'] =RES_EDU1
df_group_1['Q1'] =Q1
df_group_1
```

```python
# 평균
columns = ["gu", "J3b_3_13", "B1", "H16_2", "H16_1", "H16_5", "H16_7" ,'디지털 기기 접근성','디지털 기기 활용성','만족도']
df_group = df[columns].groupby("gu")
df_group_2 = df_group.agg('mean')
df_group_2
```

```python
columns = ["B1", "H16_2", "H16_1", "H16_5", "H16_7" ]
df_group_3 = round(df_group_2[columns])

lst = []

for i in df_group_2["J3b_3_13"]/12:
    if i < 50:
        lst.append(1)
    elif 50 <= i < 100:
        lst.append(2)
    elif 100 <= i < 150:
        lst.append(3)
    elif 150 <= i < 200:
        lst.append(4)
    elif 200 <= i < 250:
        lst.append(5)
    elif 250 <= i < 300:
        lst.append(6)
    elif 300 <= i < 350:
        lst.append(7)
    elif 350 <= i < 400:
        lst.append(8)
    elif 400 <= i < 500:
        lst.append(9)
    elif 500 <= i < 600:
        lst.append(10)
    else:
        lst.append(11)
        
df_group_3["J3b_3_13"] = lst
df_group_3
```

```python
df_group_3['만족도'] = df_group_2['만족도']
df_group_3['디지털 기기 접근성'] = df_group_2['디지털 기기 접근성']
df_group_3['디지털 기기 활용성'] = df_group_2['디지털 기기 활용성']
```

```python
df = pd.concat([df_group_1, df_group_3], axis = 1)
df = df.reset_index()
df.head()
```
